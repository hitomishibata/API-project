```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import re
```

```python
# Add underscore in front of capital letters and convert capital letters into lower cases in columns
def add_underscore_and_lowercase_columns(df):
    df.columns = df.columns.astype(str)
    for col in df.columns:
        df.rename(
            columns={col: re.sub(r"([A-Z])", r"_\1", col)}, inplace=True
            )
        
    df.columns = df.columns.str.lower()
    return df
```

```python
books_df = pd.read_parquet("books-en.parquet")
characters_df = pd.read_parquet("characters-en.parquet")
houses_df = pd.read_parquet("houses-en.parquet")
spells_df = pd.read_parquet("spells-en.parquet")
```

```python
add_underscore_and_lowercase_columns(books_df)
add_underscore_and_lowercase_columns(characters_df)
add_underscore_and_lowercase_columns(houses_df)
add_underscore_and_lowercase_columns(spells_df)
```
