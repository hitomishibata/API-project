```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import re
```

```python
#add underscore infront of capital letters

def add_underscore_and_lowercase_columns(df):
    df.columns = df.columns.astype(str)
    for col in df.columns:
        df.rename(columns={col: re.sub(r"([A-Z])", r"_\1", col)}, inplace=True)
    df.columns = df.columns.str.lower()
    return df
```

```python
books_df = pd.read_parquet("books-en.parquet")
```

```python
add_underscore_and_lowercase_columns(books_df)
```
